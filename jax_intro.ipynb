# JAX Intro
As in the [website](https://docs.jax.dev/en/latest/): JAX is a Python library for accelerator-oriented array computation and program transformation, designed for high-performance numerical computing and large-scale machine learning.

```python
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
x_jnp = jnp.linspace(0, 10, 1000)
y_jnp = 2 * jnp.sin(x_jnp) * jnp.cos(x_jnp)
plt.plot(x_jnp, y_jnp);
```
