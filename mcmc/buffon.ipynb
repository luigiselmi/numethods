# Buffon's needle
Examples from *Shonkwiler, Mendivil - Exploration in Monte Carlo Methods, 2nd Edition*

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
throws = 100000; # hash makes the rest of the line a comment
x = np.random.rand(throws) # 10000 vector of psuedorandom samples from U(0,1)
theta = 0.5 * np.pi * np.random.rand(throws) # 10000 theta’s in [0,π/2)
hits= (x<=0.5*np.sin(theta)) # vector of true/false results
print(sum(hits)) # sums and prints the number of ’true’ results
print('pi estimate:',throws/float(sum(hits))) # estimate of π
```

# Ab initio buffon needle 
Program to estimate pi by Buffon # simulation without knowing pi to start with.

```python
trials = 100000
segments = 20 # avoid division by 0
d = 2; L = 1; # for reference
pi = 1.0 # starting value for iteration
throws = 0; hits = 0
for i in range(trials): # i = 0,1,...,trials-1
    x = np.random.rand(segments) # x’s in [0,d/2)
    theta = 0.5*pi*np.random.rand(segments) # theta’s in [0,pi/2)
    hits += (x<0.5*np.sin(theta)) # unlikely to be 0 after 20 throws
    throws += segments
    pi = throws/float(sum(hits)) # cast denominator as a decimal, avoid integer division
print('estimate of pi after', throws, 'throws:', pi)
```

## Histogram

```python
throws = 100000 # 100 trials
replications=2000
y= [0.]*replications # list of 2000 zeroes
for i in range(replications): # i=0,1,...,replications-1
    x=np.random.rand(throws)
    theta=0.5*np.pi*np.random.rand(throws)
    hits= (x<=0.5*np.sin(theta))
    y[i] = (sum(hits)/float(throws)) # y= result vector
recipEst=sum(y)/float(replications)
vVec=(y-recipEst)*(y-recipEst); # vector of sqr deviations
v=sum(vVec)/(replications-1); # sample variance
stddev=np.sqrt(v) # sample standard deviation
piEst=1/recipEst
```

```python
print(piEst - math.pi)
```

```python
fig = plt.figure() # defaults to 640x480, 100 px/in
plt.hist(y,bins=100);
```
